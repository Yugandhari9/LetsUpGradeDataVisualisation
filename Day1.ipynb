## Line Chart

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.plot(x,y,'go--', linewidth=2, markersize = 12)
#go-- g means green, o means circle/marker rep, -- means how the line to be plotted(linestyle))
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
x = np.arange(1,11)
y = 3*x + 5

print(y)
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('independent')
plt.ylabel('dependent')
```

```python
#compute x and y coordinates for points on a sine curve

x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title('Sine Wave')

plt.plot(x,y)
plt.show()
```

```python
##Assignment
x = np.arange(0,10)
y = 3*x+10
plt.plot(x,y,'r*--')
plt.title('Square')
```

## SUBPLOTS

```python
plt.subplot(2,2,1) # number of rows in the plot, number of columns in the plot, every subplotid
plt.plot(x,y,'ro-')
plt.subplot(2,2,2)
plt.plot(x,y,'go-')
plt.subplot(2,2,3)
plt.plot(x,y,'b*-')
plt.subplot(2,2,4)
plt.plot(x,y,'g*-')
```
